1](a) Discrete Uniform

```python
# Discrete Uniform
# udv=Uniform discrete variable
# pd=probability Distribution
import numpy as np
import matplotlib.pyplot as plt
# for small input
u=float(input('enter upper value: '))
l=float(input('enter lower value: '))
nl=int(input('enter number of samples(large): '))
ns=int(input('enter number of samples(small): '))
udvs=np.random.uniform(l,u,ns)
pds=[1/ns]*ns
plt.bar(udvs,pds,width=0.05)
plt.plot(udvs,pds,marker='*',linestyle='',color='purple')
plt.xlabel('x')
plt.grid(True)
plt.ylabel('P(X=x)')
plt.title('Discrete Uniform PMF at small value')
plt.show()

# for big input
udvl=np.random.uniform(l,u,nl)
pdl=[1/nl]*nl
plt.bar(udvl,pdl,width=0.05)
plt.plot(udvl,pdl,marker='*',linestyle='',color='purple')
plt.xlabel('x')
plt.grid(True)
plt.ylabel('P(X=x)')
plt.title('Discrete Uniform PMF at large value')
plt.show()
```

1](b)Bernoulli

```python
# Bernoulli
import numpy as np
import matplotlib.pyplot as plt
n=int(input('number of trials: '))
# for small p
p=float(input('enter a value between 0 and 1 (small):'))
if 0<=p<=1:
  li=[0,1]
  x=np.random.choice(li, size = n)
  y=[]
  q=1-p
  for i in x:
    if i==0:
      y.append(q)
    elif i==1:
      y.append(p)
    else:
      y.append(0)
  plt.bar(x,y,width=0.05)
  plt.xlabel('x')
  plt.ylabel('P(X=x)')
  plt.grid(True)
  plt.title('Bernoulli')
  plt.show()
else:
  print('Wrong input')

# for large p
p=float(input('enter a value between 0 and 1 (large):'))
if 0<=p<=1:
  li=[0,1]
  x=np.random.choice(li, size = n)
  y=[]
  q=1-p
  for i in x:
    if i==0:
      y.append(q)
    elif i==1:
      y.append(p)
    else:
      y.append(0)
  plt.bar(x,y,width=0.05)
  plt.xlabel('x')
  plt.ylabel('P(X=x)')
  plt.grid(True)
  plt.title('Bernoulli')
  plt.show()
else:
  print('Wrong input')
```

1](c)Binomial

```python
# Binomial
def fac(n):
  m=1
  for i in range(1,n+1):
    m*=i
  return m

import numpy as np
import matplotlib.pyplot as plt
p1=float(input('enter any value between 0 and 1(large):'))
#for small p
p=float(input('enter any value between 0 and 1(small):'))
if 0<=p<=1:
  n=int(input('number of trials: '))
  k=np.arange(1,n)
  y=[]
  for i in k:
    y1=((fac(n))/((fac(i))*(fac(n-i))))*(p**i)*((1-p)**(n-i))
    y.append(y1)
  plt.plot(k,y)
  plt.bar(k,y,width=0.3,color='purple')
  plt.grid(True)
  plt.xlabel('x')
  plt.ylabel('P(X=x)')
  plt.title('Binomial PMF for small values')
  plt.show()
else:
  print('wrong input')

#for large p
if 0<=p1<=1:
  k=np.arange(1,n)
  y=[]
  for i in k:
    y1=((fac(n))/((fac(i))*(fac(n-i))))*(p1**i)*((1-p1)**(n-i))
    y.append(y1)
  plt.plot(k,y)
  plt.bar(k,y,width=0.3,color='purple')
  plt.grid(True)
  plt.xlabel('x')
  plt.ylabel('P(X=x)')
  plt.title('Binomial PMF for large values')
  plt.show()
else:
  print('wrong input')
```

1](d)Poisson

```python
# Poisson
def fac(n):
  m=1
  for i in range(1,n+1):
    m*=i
  return m

import numpy as np
import matplotlib.pyplot as plt
import math
n=int(input('enter any number: '))
lamb=float(input('enter lambda(big): '))
#for small value of lambda
lam=float(input('enter lambda(small): '))
x=np.arange(0,n)
y=[]
for i in x:
  y1=((lam**(i))*(math.exp(-lam)))/fac(i)
  y.append(y1)
plt.plot(x,y,color='purple')
plt.bar(x,y,width=0.5)
plt.grid(True)
plt.xlabel('number of events')
plt.ylabel('P(X=x)')
plt.title('Poisson PMF for small lambda')
plt.show()

#for big value of lambda

x=np.arange(0,n)
y=[]
for i in x:
  y1=((lamb**(i))*(math.exp(-lamb)))/fac(i)
  y.append(y1)
plt.plot(x,y,color='purple')
plt.bar(x,y,width=0.5)
plt.grid(True)
plt.xlabel('number of events')
plt.ylabel('P(X=x)')
plt.title('Poisson PMF for big lambda')
plt.show()
```

1](e)Geometric

```python
# Geometric
import numpy as np
import matplotlib.pyplot as plt
n=int(input('enter number of trials: '))
p=float(input('enter any value between 0 and 1(small):'))
pl=float(input('enter any value between 0 and 1(large):'))

#for small p
if 0<=p<=1:
  x=np.arange(1,n)
  y=[]
  for i in x:
    y1=((1-p)**i)*p
    y.append(y1)
  plt.plot(x,y)
  plt.grid(True)
  plt.xlabel('x')
  plt.ylabel('P(X=x)')
  plt.title('Geometric PMF (small)')
  plt.bar(x,y,color='violet',width=0.54)
  plt.show()
else:
  print('wrong input')

#for large p
if 0<=pl<=1:
  x=np.arange(1,n)
  y=[]
  for i in x:
    y1=((1-pl)**i)*pl
    y.append(y1)
  plt.plot(x,y)
  plt.grid(True)
  plt.xlabel('x')
  plt.ylabel('P(X=x)')
  plt.title('Geometric PMF(large)')
  plt.bar(x,y,color='violet',width=0.54)
  plt.show()
else:
  print('wrong input')
```

```python
# M=Number of trials
# p=probablity of success
def fac(n):
  m=1
  for i in range(1,n+1):
    m*=i
  return m

import numpy as np
import matplotlib.pyplot as plt
import math

p=float(input('enter a value between 0 and 1:'))
if 0<=p<=1:
  M=int(input('number of trials: '))
  x=np.arange(1,M)

  #for binomial pmf
  binomialpmf=[]
  for i in x:
    y=((fac(M))/((fac(i))*(fac(M-i))))*(p**i)*((1-p)**(M-i))
    binomialpmf.append(y)
  plt.plot(x,binomialpmf,label='binomial pmf')
  plt.grid(True)
  plt.xlabel('x')
  plt.ylabel('P(X=x)')
  plt.bar(x,binomialpmf,width=0.7)
  plt.title('Binomial PMF')



  #for poisson pmf
  lam=M*p
  poissonpmf=[]
  for i in x:
    y=((lam**(i))*(math.exp(-lam)))/fac(i)
    poissonpmf.append(y)
  plt.plot(x,poissonpmf,color='r',label='poisson pmf')
  plt.grid(True)
  plt.xlabel('number of events')
  plt.ylabel('P(X=x)')
  plt.title('Binomial and Poisson PMF')
  plt.bar(x,binomialpmf,width=0.3,color='r')
  plt.legend()
  plt.show()


```

Thus, it is visible from the graph that Poisson pmf and Binomial pmf are overlapping almost completely at large value of M and small value of p.Thus, binomial pmf and poisson pmf are equivalent at large value of M and small value of p. Hence Proved.
