# **Question 1**

1.[a]Uniform PDF

```python
import numpy as np
import matplotlib.pyplot as plt
a= float(input('give lower limit: '))
b= float(input('give lower limit: '))
x= np.random.uniform(0.0,2*b,1000)
x.sort()
y=[]
for i in x:
  if a<=i<=b:
    y.append(1/(b-a))
  else:
    y.append(0)
plt.plot(x,y)
# plt.ylim(0,1)
plt.grid(True)
plt.fill_between(x,y,alpha=0.4, color='g')
plt.xlabel('x')
plt.ylabel('P(X=x)')
plt.title('Uniform PDF');




```

1.[b]Gaussian/Normal PDF

```python
import numpy as np
import matplotlib.pyplot as plt
import math
s=float(input('give standard deviation: '))
mu=float(input('give mean: '))
x=np.linspace(mu-4*s,mu+4*s,1000)
y=[]
for i in x:
  y1=(1/math.sqrt(2*math.pi*s))*math.exp((-0.5*(i-mu)**2)/s**2)
  y.append(y1)
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=0.5,color='b')
plt.ylabel('P(X=x)')
plt.title('Gaussian/Normal PDF');
```

# **Question 2**

2.[a] Uniform CDF

```python
import numpy as np
import matplotlib.pyplot as plt

a= float(input('give lower limit: '))
b= float(input('give lower limit: '))
x=np.linspace(0,2*b,1000)
y=[]
for i in x:
  if a<=i<=b:
    y.append((i-a)/(b-a))
  elif i<a:
    y.append(0)
  else:
    y.append(1)
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=0.4,color='g')
plt.ylabel('P(X<=x)')
plt.title('Uniform CDF')
plt.show()

#by using cumsum

pdf=[]
for i in x:
  if a<=i<=b:
    pdf.append(1/(b-a))
  else:
    pdf.append(0)
yc=np.cumsum(pdf)/np.sum(pdf)

plt.plot(x,yc)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,yc,alpha=1,color='g')
plt.ylabel('P(X<=x)')
plt.title('Uniform CDF by cumsum');
```

2.[d] Cauchy CDF

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x0=float(input('enter x0: '))
g=float(input('enter gamma: '))
x=np.linspace(x0-5,x0+5,10000)
y=[]
for i in x:
  y1=(np.arctan((i-x0)/g))/math.pi +0.5
  y.append(y1)
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=0.5,color='c')
plt.ylabel('P(X<=x)')
plt.title('Cauchy CDF')
plt.show()


# By using cumsum


pdf=[]
for i in x:
  pdf1=(1 / (np.pi * g * (1 + ((i- x0) / g) ** 2)))
  pdf.append(pdf1)
y=np.cumsum(pdf)/np.sum(pdf)
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=1,color='c')
plt.ylabel('P(X<=x)')
plt.title('Cauchy CDF by using cumsum');
```

2.[e]Laplacian CDF

```python
import numpy as np
import matplotlib.pyplot as plt
import math

mu=float(input('enter location: '))
b=float(input('enter scale: '))
x=np.linspace(x0-7*b,x0+7*b,1000)
y=[]
for i in x:
  z=math.exp(-abs(i-x0)/b)/2
  if i<=mu:
    y.append(z)
  else:
    y.append(1-z)
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=0.5,color='r')
plt.ylabel('P(X<=x)')
plt.title('Laplacian CDF')
plt.show()

#By using cumsum
pdf=[]
for i in x:
  pdf1=(math.exp(-abs(i-mu)/b))/(2*b)
  pdf.append(pdf1)
y=np.cumsum(pdf)/np.sum(pdf)
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=1,color='r')
plt.ylabel('P(X<=x)')
plt.title('Laplacian CDF using cumsum');
```

2.[f] Rayleigh CDF

```python
import numpy as np
import matplotlib.pyplot as plt
import math

s=float(input('enter sigma(scale): '))
x=np.linspace(0,5*s,1000)
y=[]
for i in x:
  y.append(1-math.exp(-i**2/2*(s**2)))
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=0.5,color='fuchsia')
plt.ylabel('P(X<=x)')
plt.title('Rayleigh CDF')
plt.show()

#By using cumsum

pdf=[]
for i in x:
  pdf1=(i*(math.exp(-i**2/2*(s**2))))/s**2
  pdf.append(pdf1)
y=np.cumsum(pdf)/np.sum(pdf)
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=1,color='fuchsia')
plt.ylabel('P(X<=x)')
plt.title('Rayleigh CDF using cumsum');
```

# **Question 3**

3.[a] Discrete Uniform CDF

```python
import numpy as np
import matplotlib.pyplot as plt

u=float(input('enter upper value: '))
l=float(input('enter lower value: '))
ns=2000
x=np.arange(0,u+3,0.01)
y=[]
for i in x:
  if i<l:
    y.append(0)
  elif l<=i<u:
    y1=(np.floor(i)-l+1)/(u-l+1)
    y.append(y1)
  else:
    y.append(1)
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=0.5,color='coral')
plt.ylabel('P(X<=x)')
plt.title('Discrete Uniform CDF')
plt.show()
```

3.[b] Bernoulli CDF

```python
import numpy as np
import matplotlib.pyplot as plt
n=1000
p=float(input('enter a value between 0 and 1 :'))
if 0<=p<=1:
  x=np.random.uniform(-1,3,n)
  x.sort()
  y=[]
  q=1-p
  for i in x:
    if i<0:
      y.append(0)
    elif 0<=i<1:
      y.append(q)
    else:
      y.append(1)
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=0.5,color='aquamarine')
plt.ylabel('P(X<=x)')
plt.title('Bernoulli CDF')
plt.show()
```

3.[c] Binomial CDF

```python
from re import X
def fac(n):
  m=1
  for i in range(1,n+1):
    m*=i
  return m

import numpy as np
import matplotlib.pyplot as plt
p=float(input('enter any value between 0 and 1 :'))
if 0<=p<=1:
  n=int(input('number of trials: '))
  x=np.arange(1,n)
  y=[]
  for i in x:
    y1=((fac(n))/((fac(i))*(fac(n-i))))*(p**i)*((1-p)**(n-i))
    y.append(y1)
  y=np.cumsum(y)/np.sum(y)
  plt.plot(x,y)
  plt.grid(True)
  plt.xlabel('x')
  plt.fill_between(x,y,alpha=0.5,color='chocolate')
  plt.ylabel('P(X<=x)')
  plt.title('Binomial CDF')
  plt.show()
```

3.[d] Poisson CDF

```python
def fac(n):
  m=1
  for i in range(1,n+1):
    m*=i
  return m

import numpy as np
import matplotlib.pyplot as plt
import math
n=int(input('enter any number: '))
lam=float(input('enter lambda: '))
x=np.arange(0,n)
y=[]
for i in x:
  y1=((lam**(i))*(math.exp(-lam)))/fac(i)
  y.append(y1)
y=np.cumsum(y)/np.sum(y)
plt.plot(x,y)
plt.grid(True)
plt.xlabel('x')
plt.fill_between(x,y,alpha=0.5,color='darkkhaki')
plt.ylabel('P(X<=x)')
plt.title('Poisson CDF')
plt.show()
```
